1. My lane as an ML task

Task Type: Classification

2. Target or Proxy

Target: Predict whether a content page needs to be refreshed based on its performance.

3. Success Metric

Success Metric: Accuracy and F1-Score.

```python
df.head()
```

4. Unit of Analysis

One row represents one content page (article).

```python
df["trend_direction"].value_counts()
```

```python
df["needs_refresh"] = (df["trend_direction"] == "down").astype(int)

df[["trend_direction", "needs_refresh"]].head()
```

5. Why ML beats a fixed rule

A fixed rule cannot consider many factors together, such as impressions, clicks, engagement, and content age. A machine learning model can learn patterns from all these features and make better predictions.

6. Action Supported

The prediction helps the content team identify which pages should be refreshed first.
